```python
import pandas as pd
import matplotlib.pyplot as plt

FullSetClean = pd.read_csv("data_edited.csv")

stateBin = []
for row in FullSetClean["state"]:
    if row == "successful":
        stateBin.append(1)
    else:
        stateBin.append(0)

FullSetClean["StateBin"] = stateBin
FullSetClean = FullSetClean.drop(columns = "state")
```

```python
CategoryList = FullSetClean["category"].unique().tolist()

def PieChart(Category, State):
    labels = 'Successful', 'Failed'
    colors = ['lightskyblue', 'lightcoral']
    patches, texts = plt.pie(State, colors = colors)
    plt.pie(State, colors = colors, autopct = '%1.1f%%')
    plt.title('State of {} Kickstarter projects' .format(Category))
    plt.legend(patches, labels, loc="best")
    plt.show()
    
for i in CategoryList:
    Successful = 0
    Failed = 0
    tempDataFrame = FullSetClean[FullSetClean["category"] == i]
    for row in tempDataFrame["StateBin"]:
        if row == 1:
            Successful += 1
        else:
            Failed += 1
    PieChart(i, State = [Successful, Failed])
```
